#### log scale , scales our data so its easier to understand

```python
fig, ax = plt.subplots()
_ = ax.set_title('Titanic Data Set')
_ = ax.plot(np.sort(df.fare), marker = 'o',linestyle='', markersize=5)
_ = ax.set_ylabel('fare')
_ = ax.set_xlabel('people')
_ = ax.set_yscale('log')
_ = ax.set_yticks([3,12,30,870])
_ = ax.set_yticklabels(['3rd','2nd','1stb','1sta'])
```
