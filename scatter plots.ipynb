## import stuff

```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

## Load data

```python
df = pd.read_csv('titanic-dataset.csv').dropna()
```

```python
df.head()
```

## scatter plot

```python
fig,ax = plt.subplots()
_ = ax.scatter(df['age'], df['fare'])
_ = ax.set_xlabel('age')
_ = ax.set_ylabel('fare')
```

```python
fig,ax = plt.subplots()
_ = ax.scatter('age', 'fare',data=df)
_ = ax.set_xlabel('age')
_ = ax.set_ylabel('fare')
```

```python
fig,ax = plt.subplots()
_ = ax.scatter('age', 'fare',data=df,color='hotpink')
_ = ax.set(xlabel='age',ylabel='fare',title='age vs fare')
```

### Lived vs Died

```python
fig,ax = plt.subplots()
_ = ax.scatter('age', 'fare',data=df,c='survived')
_ = ax.set(xlabel='age',ylabel='fare',title='age vs fare')
```

## color map

```python
fig,ax = plt.subplots()
_ = ax.scatter('age', 'fare',data=df,c='survived',cmap='tab10')
_ = ax.set(xlabel='age',ylabel='fare',title='age vs fare')
```

## lets make our own colormap

```python
import matplotlib.colors as mcolors
cm = mcolors.ListedColormap(['black','red'])
fig,ax = plt.subplots()
_ = ax.scatter('age', 'fare',data=df,c='survived',cmap=cm)
_ = ax.set(xlabel='age',ylabel='fare',title='age vs fare')
```

```python
fig,ax = plt.subplots()
sc = ax.scatter('age', 'fare',data=df,c='survived',cmap=cm)
_ = ax.set(xlabel='age',ylabel='fare',title='age vs fare')
_ = ax.legend(*sc.legend_elements())
```

```python
fig,ax = plt.subplots()
sc = ax.scatter('age', 'fare',data=df,c='survived',cmap=cm)
_ = ax.set(xlabel='age',ylabel='fare',title='age vs fare')
_ = ax.legend(handles = sc.legend_elements()[0],
             labels = ['Died','Survived'])
```

```python
fig,ax = plt.subplots()
sc = ax.scatter('age', 'fare',data=df,c='survived',
                cmap=cm,alpha=0.5)
_ = ax.set(xlabel='age',ylabel='fare',title='age vs fare')
_ = ax.legend(handles = sc.legend_elements()[0],
             labels = ['Died','Survived'])
```

```python
df.pclass.value_counts()
```

```python
fig,ax = plt.subplots()
sc = ax.scatter('age', 'fare',data=df,c='survived',
                cmap=cm,s='pclass')
_ = ax.set(xlabel='age',ylabel='fare',title='age vs fare')
_ = ax.legend(handles = sc.legend_elements()[0],
             labels = ['Died','Survived'])
```

```python
fig,ax = plt.subplots()
sc = ax.scatter('age', 'fare',data=df,c='survived',
                cmap=cm,s=df['pclass']*75,
               alpha=0.5)
_ = ax.set(xlabel='age',ylabel='fare',title='age vs fare')
_ = ax.legend(handles = sc.legend_elements()[0],
             labels = ['Died','Survived'])
```
