# sin curve

```python
x = np.linspace(0,2*np.pi,100)
sinx = np.sin(x)
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x,sinx)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
_ = ax.plot(x,sinx,marker='*',linestyle='')
_ = ax.set_aspect('equal') # sets aspect ration
```

# change color
color : changes the color of the prominent object in the plot. eg in line graph, line is the prominent object.

```python
fig, ax = plt.subplots(figsize=(15,10))
_ = ax.plot(x,sinx,marker='.',linestyle='',markeredgecolor='green',
           markerfacecolor='white',markersize=20)
_ = ax.set_aspect('equal') # sets aspect ration
```
