# Import stuff

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

# Load dataset

```python
df = pd.read_csv('titanic-dataset.csv')
```

### Super title (figure title)

```python
fig, (ax1, ax2) = plt.subplots(ncols=2)
_ = fig.suptitle('2 Axes')
```

# Grid

```python
fig, axes = plt.subplots(ncols=2,nrows=2,constrained_layout=True)
_ = fig.suptitle('Grid')
```

## making 2 axes

```python
fig, (ax1, ax2) = plt.subplots(ncols=2)
_ = fig.suptitle('Titanic Data set')
_ = ax1.plot(np.sort(df.age),marker = 'o', markersize=1)
_ = ax1.set_ylabel('Ages')
_ = ax2.plot(np.sort(df.fare), marker='o', markersize=1)
_ = ax2.set_ylabel('Fare')
```

## Constrained layout

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,constrained_layout=True)
_ = fig.suptitle('Titanic Data set')
_ = ax1.plot(np.sort(df.age),marker = 'o', markersize=1)
_ = ax1.set_ylabel('Ages')
_ = ax2.plot(np.sort(df.fare), marker='o', markersize=1)
_ = ax2.set_ylabel('Fare')
```

## Grid Spec API

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               gridspec_kw={'width_ratios':[1,3]},
                               constrained_layout=True)
_ = fig.suptitle('Titanic Data set')
_ = ax1.plot(np.sort(df.age),marker = 'o', markersize=1)
_ = ax1.set_ylabel('Ages')
_ = ax2.plot(np.sort(df.fare), marker='o', markersize=1)
_ = ax2.set_ylabel('Fare')
```
