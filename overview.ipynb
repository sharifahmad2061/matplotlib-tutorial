# Read the titanic data

```python
df = pd.read_csv('titanic-dataset.csv')
```

```python
fig,ax = plt.subplots()
_ = ax.set_title('Titanic Data Set')
_ = ax.set_ylabel('Ages')
_ = ax.set_xlabel('Passengers')
_ = ax.plot(np.sort(df.age),marker='o',markersize=1)
plt.show()
```
