### Setup data for plot

```python
x = np.linspace(0,2*np.pi,100)
cosx = np.cos(x)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
_ = ax.plot(x,cosx,linestyle=':',linewidth=20,
            color='purple',label='cosine')
_ = ax.set_aspect('equal') # sets aspect ration
_ = ax.legend()
```
